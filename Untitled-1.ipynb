```python
# 2) Data cleaning
clean_df = df.copy()

# Drop duplicate employee records if any
before = clean_df.shape[0]
clean_df = clean_df.drop_duplicates(subset=['employee_id'])
after = clean_df.shape[0]
print(f'Dropped duplicate employee records: {before - after}')

# Standardize string values
for col in clean_df.select_dtypes(include='object').columns:
    clean_df[col] = clean_df[col].astype(str).str.strip()

# Missing value handling
num_cols = clean_df.select_dtypes(include='number').columns.tolist()
cat_cols = clean_df.select_dtypes(exclude='number').columns.tolist()

# Handle monthly_income by job_role and job_level
if 'monthly_income' in num_cols and clean_df['monthly_income'].isnull().any():
    # Create a mapping of group medians
    group_median = clean_df.groupby(['job_role', 'job_level'])['monthly_income'].median()
    
    # Impute nulls using the group median
    null_mask = clean_df['monthly_income'].isnull()
    for idx in clean_df[null_mask].index:
        job_role = clean_df.loc[idx, 'job_role']
        job_level = clean_df.loc[idx, 'job_level']
        group_key = (job_role, job_level)
        if group_key in group_median.index:
            clean_df.loc[idx, 'monthly_income'] = group_median[group_key]
    
    # For any remaining nulls, use overall median
    if clean_df['monthly_income'].isnull().any():
        clean_df['monthly_income'].fillna(clean_df['monthly_income'].median(), inplace=True)

# Handle other numeric columns with overall median
for col in num_cols:
    if col != 'monthly_income' and clean_df[col].isnull().any():
        clean_df[col] = clean_df[col].fillna(clean_df[col].median())

for col in cat_cols:
    if clean_df[col].isnull().any():
        clean_df[col] = clean_df[col].fillna(clean_df[col].mode().iloc[0])

# Target encoding for analysis
clean_df['attrition_flag'] = (clean_df['attrition'].str.lower() == 'yes').astype(int)

print('Remaining missing values:', int(clean_df.isnull().sum().sum()))
print('Cleaned shape:', clean_df.shape)
display(clean_df.head())
```

```python
# 2a) Analyze job_satisfaction patterns with key factors (before cleaning nulls)
print('=== Job Satisfaction Analysis (using non-null values) ===')
non_null_sat = df[df['job_satisfaction'].notna()].copy()

# Key factors influencing satisfaction
satisfaction_factors = ['percent_salary_hike', 'bonus_eligible', 'overtime_paid_eligible', 
                        'work_life_balance', 'environment_satisfaction', 'overtime_hours_per_week', 
                        'skills_growth_opportunities']

print(f'\nNon-null job_satisfaction records: {len(non_null_sat)} / {len(df)}')
print(f'Average job_satisfaction (non-null): {non_null_sat["job_satisfaction"].mean():.2f}')

# Analyze correlation with satisfaction
print('\nCorrelation of factors with job_satisfaction:')
for factor in satisfaction_factors:
    if factor in non_null_sat.columns:
        corr_val = non_null_sat['job_satisfaction'].corr(non_null_sat[factor])
        print(f'  {factor}: {corr_val:.3f}')

# Analyze satisfaction by bonus_eligible and overtime_paid_eligible
print('\nAverage satisfaction by bonus_eligible & overtime_paid_eligible:')
bonus_overtime_analysis = non_null_sat.groupby(['bonus_eligible', 'overtime_paid_eligible'])['job_satisfaction'].agg(['mean', 'count']).round(2)
display(bonus_overtime_analysis)
```

```python
# Verify group-based imputation for monthly_income
print('=== Monthly Income Imputation by Job Role & Job Level ===')
print('\nMedian monthly income by job_role and job_level:')
income_medians = clean_df.groupby(['job_role', 'job_level'])['monthly_income'].agg(['median', 'count', 'mean']).round(2)
display(income_medians)

print(f'\nMonthly income missing values after group imputation: {clean_df["monthly_income"].isnull().sum()}')
print(f'\nRemaining missing values by column:')
display(clean_df.isnull().sum()[clean_df.isnull().sum() > 0].sort_values(ascending=False))
```

```python
# Diagnostic: Check which job_role/job_level combinations have ALL null monthly_income
print('=== Groups with NULL monthly_income values ===')
null_mask = clean_df['monthly_income'].isnull()
if null_mask.any():
    null_by_group = clean_df[null_mask].groupby(['job_role', 'job_level']).size()
    print(f'Groups with null values: {len(null_by_group)}')
    display(null_by_group)
    
    # Check if these groups have non-null values elsewhere
    print('\nNon-null counts by these groups:')
    for (job_role, job_level) in null_by_group.index:
        group_data = clean_df[(clean_df['job_role'] == job_role) & (clean_df['job_level'] == job_level)]
        non_null_count = group_data['monthly_income'].notna().sum()
        total_count = len(group_data)
        print(f'  {job_role} (Level {job_level}): {non_null_count} non-null / {total_count} total')
```

```python
# 4) Feature engineering
feat_df = clean_df.copy()

# Ratio features
feat_df['tenure_ratio'] = feat_df['years_in_current_role'] / (feat_df['years_at_company'] + 1)
feat_df['income_per_experience_year'] = feat_df['monthly_income'] / (feat_df['total_working_years'] + 1)

# Binary behavior flags
feat_df['is_frequent_traveler'] = feat_df['business_travel'].isin(['Travel_Frequently']).astype(int)
feat_df['is_remote_heavy'] = (feat_df['remote_work_days_per_week'] >= 3).astype(int)

# Bucketized distance feature
feat_df['distance_bucket'] = pd.cut(
    feat_df['distance_from_home_km'],
    bins=[-1, 5, 15, 30, np.inf],
    labels=['Very_Near', 'Near', 'Moderate', 'Far']
)

print('Feature engineered dataset shape:', feat_df.shape)
display(feat_df[['employee_id', 'tenure_ratio', 'income_per_experience_year', 'is_frequent_traveler', 'is_remote_heavy', 'distance_bucket']].head())
```

```python
# 5) Hypothesis testing
alpha = 0.05

# -------------------------------------------------
# Hypothesis 1: Two-proportion Z test
# H0: Attrition rate is the same for overtime paid eligible Yes and No employees.
# H1: Attrition rates are different between the two groups.
# -------------------------------------------------
g1 = feat_df[feat_df['overtime_paid_eligible'] == 'Yes']
g2 = feat_df[feat_df['overtime_paid_eligible'] == 'No']

x1, n1 = g1['attrition_flag'].sum(), g1.shape[0]
x2, n2 = g2['attrition_flag'].sum(), g2.shape[0]

p1, p2 = x1 / n1, x2 / n2
p_pool = (x1 + x2) / (n1 + n2)
se = np.sqrt(p_pool * (1 - p_pool) * (1 / n1 + 1 / n2))
z_stat = (p1 - p2) / se
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

print('Z-Test (Two Proportions):')
print(f'  Attrition rate (Overtime Eligible=Yes): {p1:.4f}')
print(f'  Attrition rate (Overtime Eligible=No):  {p2:.4f}')
print(f'  z-statistic = {z_stat:.4f}, p-value = {p_value_z:.6f}')
print('  Decision:', 'Reject H0' if p_value_z < alpha else 'Fail to reject H0')
print()

# -------------------------------------------------
# Hypothesis 2: Independent two-sample T test (Welch)
# H0: Mean monthly income is equal for attrition Yes and No groups.
# H1: Mean monthly income differs between attrition Yes and No groups.
# -------------------------------------------------
income_yes = feat_df.loc[feat_df['attrition'] == 'Yes', 'monthly_income']
income_no = feat_df.loc[feat_df['attrition'] == 'No', 'monthly_income']

t_stat, p_value_t = ttest_ind(income_yes, income_no, equal_var=False)

print('T-Test (Welch, Monthly Income by Attrition):')
print(f'  Mean income (Attrition=Yes): {income_yes.mean():.2f}')
print(f'  Mean income (Attrition=No):  {income_no.mean():.2f}')
print(f'  t-statistic = {t_stat:.4f}, p-value = {p_value_t:.6f}')
print('  Decision:', 'Reject H0' if p_value_t < alpha else 'Fail to reject H0')
print()

# -------------------------------------------------
# Hypothesis 3: Chi-square test of independence
# H0: Attrition is independent of business travel frequency.
# H1: Attrition depends on business travel frequency.
# -------------------------------------------------
cont_table = pd.crosstab(feat_df['business_travel'], feat_df['attrition'])
chi2_stat, p_value_chi, dof, expected = chi2_contingency(cont_table)

print('Chi-Square Test (Business Travel vs Attrition):')
print('Contingency Table:')
display(cont_table)
print(f'  chi2 = {chi2_stat:.4f}, dof = {dof}, p-value = {p_value_chi:.6f}')
print('  Decision:', 'Reject H0' if p_value_chi < alpha else 'Fail to reject H0')
```
